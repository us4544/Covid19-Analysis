```python
figure(figsize=(15,8), dpi=120, facecolor='w', edgecolor='k')
plt.barh(total_states,df['positive'], align='center', alpha=.7,color=(1,0,0),edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['positive'])) 
plt.xlabel('Positive Cases')  
plt.ylabel('States')
plt.title('Covid Cases')  
plt.show()
```

```python
figure(figsize=(10,8), dpi=120, facecolor='w', edgecolor='k')
plt.barh(total_states,df['new_active'], align='center', alpha=1, color=(1,0.5,0),edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, df['state_name'])  
plt.xlim(2,max(df['new_active'])) 
plt.xlabel('Active Cases')  
plt.ylabel('States')
plt.title('Covid Cases')  
plt.show()
```

```python
figure(figsize=(10,8), dpi=120, facecolor='w', edgecolor='k')
plt.barh(total_states,df['death'], align='center', alpha=0.5, color=(0,.5,1),  edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['death'])) 
plt.xlabel('Death')  
plt.ylabel('States')  
plt.title('Corona Virus Cases') 
plt.show()
```
